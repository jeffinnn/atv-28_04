```python
cliente = input(f"Qual é o seu nome completo? \n")
temp = float((input(f"\nQual é a sua temperatura? \n")).replace("," , "."))
pressao = int(input(f"\nQual é a sua pressão? \n"))
dor = int(input(f"\nDe 0 a 10 qual é o nível da sua dor? \n "))

if dor > 10 :
  print(f"\nFale sua dor somente de 0 a 10 , Tente novamente")

if temp > 39.5 or pressao > 180 or dor >=9:
    classificacao = "Emergencia"
elif temp > 38.5 or pressao > 160 or dor >= 7:
      classificacao = "Muito urgente"
elif temp > 37.5 or pressao > 140 or dor >= 4:
    classificacao = "Urgente"
else:
    classificacao = "Pouco urgente"

print(f"\n {cliente}")
print(f"\nSua temperatura é: {temp}")
print(f"Sua pressão é: {pressao}")
print(f"Sua dor é: {dor}")
print(f"\nSeu situação é : {classificacao}")
```
